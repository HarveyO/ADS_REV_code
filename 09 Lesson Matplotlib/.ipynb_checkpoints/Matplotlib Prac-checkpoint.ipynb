# Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
%matplotlib
import matplotlib.pyplot as plt
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

## Exercise 2

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(x,y)
plt.show()
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
plt.show()
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4],xlim =(20,22),ylim =(30,50))

ax1.plot(x,z)
ax2.plot(x,y)

ax2.set_xlabel('x')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

ax1.set_xlabel('x')
ax1.set_ylabel('Z')

plt.show()

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4],xlim =(20,22),ylim =(30,50))

ax1.plot(x,z)
ax2.plot(x,y)

ax2.set_xlabel('x')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

ax1.set_xlabel('x')
ax1.set_ylabel('Z')

plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
ax1 = plt.subplots(nrows=1, ncols=2)

plt.show()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.plot(x,y,  linewidth=3, linestyle='dashed', color='purple')
ax2.plot(x,z, linewidth=5, color='g')
plt.show()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig = plt.figure(figsize=(12,2))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x,y,  linewidth=5, color='b')
ax2.plot(x,z, linewidth=3, color='r', linestyle='dashed')

ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.set_xlabel('x')
ax2.set_ylabel('z')
plt.show()
```
